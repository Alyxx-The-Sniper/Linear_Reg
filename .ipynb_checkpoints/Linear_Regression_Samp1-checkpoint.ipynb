## Check Categorical Features

For simplicity lets drop some categories that has small observation number (Pike, Smelt, Parkki, Whitefish).

```python
df.Species.value_counts()
```

```python
# Subset to just Perch, Bream, and Roach species only
df = df[df.Species.isin(['Perch', 'Bream', 'Roach'])]
# verify
print(df.Species.unique())
df.Species.value_counts()
```

## Perform one-hot encoding (OHE) on categorical data
`Note:` 
<br/>
 - OHE is used in this case due to the presence of nominal categories without a specific order.
 - For ordinal categories, alternatives like mapping, label encoding, or ordinal encoding can be employed.
 - If the resulting OHE representation yields an excessive number of columns, it is possible to simplify the categories by grouping them into regions (e.g., ASIA, EUROPE) or explore the use of Binary Encoder as an alternative.



```python
# create one hot encoder
ohe = pd.get_dummies(df['Species'])
```

```python
# combine ohe to our df in a new dataset df1
df1 = df.join(ohe)
```

```python
# Drop Species column
# preview our new dataset
df1 = df1.drop('Species', axis=1)
df1.head(2)
```

## Correaltion Heatmap
Check for multicolinearity, We will drop out target features (weight) to easily identify multicolinearity features.
To check for multicollinearity, we will exclude the target feature (weight) in order to more effectively identify the features that may exhibit multicollinearity.

```python
ax2= plt.figure(figsize=(8,5))
ax2=sns.heatmap(df.drop('Weight', axis=1).corr(), annot=True, fmt='.3', cmap='RdBu_r')
plt.title('Features Heatmap')
ax2=plt.show()
```

## Variance Inflation Factor (VIF) 
The Variance Inflation Factor (VIF) is a statistical measure used to quantify the extent of multicollinearity in a regression analysis. Multicollinearity occurs when two or more predictor variables in a regression model are highly correlated with each other.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
features = df.drop(['Weight', 'Species'], axis=1 )
features.info()
```

```python
vif = pd.DataFrame()
vif["Feature"] = features.columns
vif["VIF"] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]

# Identify features with high VIF indicating multicollinearity
multicollinear_features = vif[vif["VIF"] > 1]  # Adjust the threshold as needed
```

```python
print("Multicollinear Features:")
print(multicollinear_features)
```

We will keep the features with the lowest VIF values, which indicates a lower degree of multicollinearity. Additionally, we will select the feature that shows the highest correlation with our target feature (weight)

## Correlation between features and target (Weight)

```python
# 
plt.figure(figsize=(15, 5)) 
corr_matrix = df.corr()
 
# # Isolate the column corresponding to `Winnings`
corr_target = corr_matrix[['Weight']].drop(labels=['Weight'])
corr_target = corr_target.sort_values(by=['Weight'], ascending=False)

ax1=plt.subplot(1,2,1)
ax1=sns.heatmap(corr_target, annot=True, fmt='.3', cmap='RdBu_r')
plt.title('Features correalation to weight')
plt.show()
```

There is a significant presence of multicollinearity among our features. 
<br/>
We have two options: 
1. We can either keep all the features as they are 
2. Or we can selectively drop some features that shows significant degree of multicollinearity to each other and weaker correlation with our target variable (weight).

The main goal of Linear Regression is to predict an outcome using individual features. In this activity, our objective is to identify at least two features for our regression model. To achieve this, we will prioritize features with the lowest VIF values, indicating a lower level of multicollinearity. Additionally, we will consider another feature that demonstrates the highest correlation with our target variable.

Based on our analysis, the features `Diagonal_Length` and `Height` meet these criteria and can be selected as potential features for our Linear Regression model.

# Linear Regresion Model

#### Define X and y

```python
y = df1['Weight']
# Selected features
X = df1[['Diagonal_Length', 'Height']]
# All features
# X = df1.drop(columns=['Weight', 'Species'])
```

#### Train Test split
Spliting our data to training 70% and testing 30%.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size = 0.30, random_state = 0)
```

#### Normalize the Data
Our features(X) must be in the same scale

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train[:5]
```

```python
 X_test[:5]
```

```python
X_train[:5]
```

```python
 X_test[:5]
```

#### Intialize Linear  Regression Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

#### Fit the Model

```python
lr.fit(X_train, y_train)
```

#### Model Accuracy 
R2 score
<br/>
r2 = [variation of the mean - var(reg line)]  / var(mean)

```python
lr.score(X_train,y_train), lr.score(X_test,y_test)
```

## Visualize Train and Test data

```python
#
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
###########################################
plt.figure(figsize=(15, 5)) 

# Test dataset
ax1 = plt.subplot(1,2,2)
ax1=sns.regplot(lr.predict(X_test), y_test,
               scatter_kws={"color": "black"}, line_kws={"color": "red"}
               )
ax1=plt.xlabel('Predicted Weight')
ax1=plt.ylabel('Actual Weight')
ax1=plt.title('Test Dataset')

# Train dataset
ax2 = plt.subplot(1,2,1)
ax2=sns.regplot(lr.predict(X_train), y_train,
               scatter_kws={"color": "black"}, line_kws={"color": "red"}
               )
ax2=plt.xlabel('Predicted Weight')
ax2=plt.ylabel('Actual Weight')
ax2=plt.title('Train Dataset')
plt.show()
```

## Actual vs. Predicted Weight

```python
y_predict = lr.predict(X_test)
```

```python
plt.figure(figsize=(5, 4)) 

sns.regplot(y_test, y_predict)
plt.title("Actual Vs. Predicted Weight")
plt.xlabel("Actual Weight")
plt.ylabel("Predicted Weight")
plt.show()
```

## Sample Prediction

```python
def Predict_Weight(Diagonal_Length, Height):
    new_fish = [[Diagonal_Length, Height]]
    new_fish = scaler.transform(new_fish)
    new_fish_predict = lr.predict(new_fish)
    
    return new_fish_predict    
```
